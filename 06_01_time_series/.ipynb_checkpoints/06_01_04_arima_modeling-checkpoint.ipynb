## ARIMA Modeling
- One more key branch of time series modeling, which has to do with random movement (rather than seasonality)
- Time series that are observed at regular intervals
    - Consistency matters more than what the actual interval is
- Simplest way to approach the evolution of a process is with a single trend term
    - Typically implies some linear relationship: as time moves forward, outcome changes at a given rate per unit of time
    - Sometimes want to model temporal evolution saying that a certain set of previous values influence the next outcome
        - Look back a certain distance to see how to look forward
        - Allows to adjust to shocks or changes to the system
        - Lets outliers fade from influence over time
        - Incorporates the underlying effects within the randomness
- **ARIMA (AutoRegressive Integrated Moving Average)** and variations ARMA(AutoRegressive Moving Average), ARIMAx( ARIMA with Explanatory Variable)
    - Fundamental technique(s) for the second case above
    - Extension of least squares regression
    - Allows for *auto regressions* and *moving averages* to be added to specifications
    - Generally apply to a stationary process
### Stationarity and Differencing
- **Differencing**: subtracting each term from its following term
    - If there exists a constant linear trend, the difference has a constant expectation
    - Extendable: if the rate of change is increasing at a constant rate, the difference in differences yields a stationary time series
- Goal is to end up with a stationary process
    - Stationary process doesn't mean expected value of difference is zero
    - Instead, may have a constant expected value (drift of the original process)
### Autoregressive Models
- **Autoregressive model**: a linear model where the terms of interest are the previous values of the outcome
    - Simple example: today's outcome is based on yesterday's outcome and some random noise term
$$X_{t}=\alpha+\beta X_{t-1}+\epsilon$$
    - $\alpha =$ constant or adjustment factor
    - $\beta = $ the amount the previous value influence the next value
    - $\epsilon =$ noise

```python
# simulate an autoregressive model
x_ar = [15]
for i in range(500):
    y = 5 + .7*x_ar[i-1]
    x_ar.append(float(y))
```

```python
plt.plot(x_ar)
plt.title('autoregressive example')
plt.show()
```

- Random movement generated from an autoregressive model
- Autoregressive models are generalized to AR(n)
    - This model is AR(1): only regressing on one of the previous terms (the first one)
    - n is the number of autoregressive terms included (also called order)
    - Generally speaking, autoregressive models don't skip terms
    - AR(n) implies that the terms are the lags of 1 through n
    - Movement comes from the random error (in this model, standard normal)


```python
# change the value of beta to 1
x_ar = [15]
for i in range(500):
    y = 5 + 1*x_ar[i-1]
    x_ar.append(float(y))

plt.figure(figsize=(8,8))
plt.subplot(221)
plt.plot(x_ar)
plt.title('alpha =  5, beta = 1')

# change the value of alpha to 0 beta to 1
x_ar = [15]
for i in range(500):
    y = 0 + 1*x_ar[i-1]
    x_ar.append(float(y))

plt.subplot(222)
plt.plot(x_ar)
plt.title('alpha = 0, beta = 1')

# change the value of beta to >1
x_ar = [15]
for i in range(500):
    y = 5 + 2*x_ar[i-1]
    x_ar.append(float(y))

plt.subplot(223)
plt.plot(x_ar)
plt.title('alpha = 5, beta = 2')

x_ar = [15]
for i in range(500):
    y = 1 + 5*x_ar[i-1]
    x_ar.append(float(y))
    
plt.subplot(224)
plt.plot(x_ar)
plt.title('alpha = 1, beta = 5')

plt.show()
```

### Moving Average
- Shift focus from previous outcome to the previous error

```python
x_ma = [15]
e_t = 0
for i in range(500):
    e = np.random.randn(1)
    y = 1 + .3*e_t + e
    x_ma.append(float(y))
    e_t=e

plt.plot(x_ma)
plt.title('moving average example')
plt.show()
```

```python
print((x_ma[1:21]))
```

- Moving average centered around 5
- Moving average model: $X_t=\alpha + \beta \epsilon _{t-1}+\epsilon _t$
- Model is MA(1): like with AR(n), MA(n) can make next term dependent on as many previous errors necessary

```python
# beta = 1
x_ma = [15]
e_t = 0
for i in range(500):
    e = np.random.randn(1)
    y = 1 + 1*e_t + e
    x_ma.append(float(y))
    e_t=e

plt.figure(figsize=(8,8))
plt.subplot(221)
plt.plot(x_ma)
plt.title('alpha = 1, beta = 1')

# change the value of alpha to 0 beta to 1
x_ma = [15]
e_t = 0
for i in range(500):
    e = np.random.randn(1)
    y = 0 + 1*e_t + e
    x_ma.append(float(y))
    e_t=e

plt.subplot(222)
plt.plot(x_ma)
plt.title('alpha = 0, beta = 1')

# change the value of beta to >1
x_ma = [15]
e_t = 0
for i in range(500):
    e = np.random.randn(1)
    y = 1 + 2*e_t + e
    x_ma.append(float(y))
    e_t=e
    
plt.subplot(223)
plt.plot(x_ma)
plt.title('alpha = 1, beta = 2')

x_ma = [15]
e_t = 0
for i in range(500):
    e = np.random.randn(1)
    y = 1 + 5*e_t + e
    x_ma.append(float(y))
    e_t=e

plt.subplot(224)
plt.plot(x_ma)
plt.title('alpha = 1, beta = 5')

plt.show()
```

### ARMA Modeling
- Combine the two classes of coefficient into a single model to create an ARMA model
- ARMA model defined in two terms:
    - $p=$ order of the autoregressive component
    - $q=$ order of the moving average component
    - Written as ARMA(p,q)
    - ARMA(1,1): $X_t=\alpha +\beta _{1}X_{t-1}+\beta_2\epsilon_{t-1}+\epsilon_t$

### Finding the Order
- For ARIMA, using statsmodels to implement and analyze the datasets
    - Sklearn requires you rederive columns for ar and ma terms yourself
- Using a partial autocorrelation function
    - Takes the correlation of time series with a lagged version of itself
    - Each bar represents the correlation with another lag
    - pacf at 1 is the correlation of initial series with itself lagged by one
    - For these plots, 0 will always return 1 since the time series perfectly correlates with itself

```python
# reset x_ar and x_ma
x_ar = [15]
for i in range(500):
    y = 5 + .7*x_ar[i-1]
    x_ar.append(float(y))
    
x_ma = [15]
e_t = 0
for i in range(500):
    e = np.random.randn(1)
    y = 1 + .3*e_t + e
    x_ma.append(float(y))
    e_t=e
    
from statsmodels.tsa.stattools import acf, pacf
x_acf = pd.DataFrame(pacf(x_ar))

x_acf.plot(kind='bar')
plt.title('AR(1) PACF')
plt.show()
```

```python
x_acf = pd.DataFrame(pacf(x_ma))
x_acf.plot(kind='bar')
plt.title('MA(1) PACF')
plt.show()
```

- Plot interpretation: want to see where values are outliers rather than looking for a specific value
    - Indicates a possibly strong autocorrelation
    - AR plot: 2 is strong autocorrelation
    - MA plot: 1 is strong autocorrelation
    - Look at p of 2 and 1 respectively for models
    
**If all values are high**
- Likely means need to use differences
    - Can manually create differences
    - Implement the I aspect using ARIMA models
        - I: level of differencing
        - 1 is simple difference, 2 is difference of differences, etc
        - Can use high degrees of differencing, but need to make sure theres a logical reason to
        - High degrees can look like they work when really there's nothing in the process making high degree of I logical
- Other techniques to find these parameters
    - Grid search
    - Trial and error: attempt several reasonable combinations of p and q, see what matches data best
    - [Information criteria](https://coolstatsblog.com/2013/08/14/using-aic-to-test-arima-models-2/)
