# Stationarity

### Prerequisite concepts
- **Stochastic process**: a sequence of random variables understood to have happened over a course of time
    - At every new unit of time, variables can assume one of many possible values
    - Each value has a probability associated with it
    - Cannot know the exact values a variable will take
    - Instead can make inferences based on the aforementioned probabilities
    - **Bernoulli process**: set of iid (independent and identically distributed) variables following a Bernoulli distribution
        - Simplest form of a stochastic process
        - Independent: outcome of one observation does not affect other outcomes
        - Identically distributed: each variable is drawn from the same distribution
- **Random Walks**: a sequence of random variables
    - Random variables are iid
    - **Simple random walk** 
        - Size of each step in the sequence is fixed (equal to 1)
        - Only the direction is random
    - **Symmetric simple random walk**:
        - Probability of step is 0.5

### Stationarity
- Stationary vs drift
    - **Stationary**: process where future value is expected to be where the process is currently (expected value is 0-
    - **Drift**: process that is expected to deviate from its current position
- A time series has **stationarity** if a shift in time doesn't cause a change in the shape of the distribution
    - Properties of the distribution such as mean, variance, covariance are constant over time
    - Series is without seasonality and trend
- Intuitively, a stationary process is one that does not evolve statistically
    - Example: placing a drop of ink into a glass of water
        - Immediately after the drop of ink, the water is clear with a small black speck
        - Ink starts to diffuse and the speck grows
        - This will continue until the whole glass is a uniform black color
        - Once the ink has permeated the entire glass of water, nothing much else will happen
        - The ink will never spontaneously return to its initial state
    - The ink particles behave as **random walks**
        - Each particle wanders off in its own direction
        - Cannot get more random once the ink is distributed uniformly throughout the glass

### Types of stationarity
- Strict stationarity
    - Series where joint distribution of any moments of any degree within the process is never dependent on time
    - In practice, this definition is too strict to be used for any real-life model
- First-order stationarity
    - Series where means that never change with time
    - Any other statistic (e.g. variance) can change
- Second-order stationarity (also called weak stationarity)
    - Series mean, variance, and autovariance are constant over time
    - Other statistics can change over time
    - Constrained version of strict stationarity, very common
- Trend-stationarity
    - Models fluctuate around a deterministic trend (the series mean)
    - Amplitude (height of one oscillation) is constant across the series
- Difference-stationarity
    - Models that need one or more differencings to become stationary

### Determining stationarity
- Can't usually determine a model is stationary or not by looking at a graph
- Hypothesis tests for stationarity
    - Run sequence plot
    - Unit root tests
        - Augmented Dickey-Fuller (ADF) test
        - Zivot-Andrews test
    - KPSS test: run as a complement to the unit  root tests
    - Spectrum analysis based tests: less common tests
        - Priestly-Subba Rao (PSR) test
        - Wavelet-based test
